```python
 def train_predict():
    n=int(input('Please Enter Number of data points you want to predict'))  #Here we are asking how many datapoints to expect from user to predict
    data=pd.DataFrame(columns=['orientation_X','orientation_Y','orientation_Z','orientation_W','angular_velocity_X','angular_velocity_Y','angular_velocity_Z','linear_acceleration_X','linear_acceleration_Y','linear_acceleration_Z','series_id'])
    for i in range(n):
            print('Enter values for datapoint number {}'.format(i+1))
            samples= 128
            df=pd.DataFrame()
            df['orientation_X'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_X : ").split(', ')][:samples] # input for each features and storing it in dataframe
            df['orientation_X'] = pd.to_numeric(df['orientation_X'])
            df['orientation_Y'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_Y :").split(', ')][:samples]
            df['orientation_Y'] = pd.to_numeric(df['orientation_Y'])
            df['orientation_Z'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_Z :").split(', ')][:samples]
            df['orientation_Z'] = pd.to_numeric(df['orientation_Z'])
            df['orientation_W'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_W :").split(', ')][:samples]
            df['orientation_W'] = pd.to_numeric(df['orientation_W'])

            df['angular_velocity_X'] = [float(x) for x in input("\nEnter 128 samples for angular_velocity_X : ").split(', ')][:samples]
            df['angular_velocity_X'] = pd.to_numeric(df['angular_velocity_X'])
            df['angular_velocity_Y'] = [float(x) for x in input("\nEnter 128 samples for angular_velocity_Y : ").split(', ')][:samples]
            df['angular_velocity_Y'] = pd.to_numeric(df['angular_velocity_Y'])
            df['angular_velocity_Z'] = [float(x) for x in input("\nEnter 128 samples for angular_velocity_Z : ").split(', ')][:samples]
            df['angular_velocity_Z'] = pd.to_numeric(df['angular_velocity_Z'])

            df['linear_acceleration_X'] = [float(x) for x in input("\nEnter 128 samples for linear_acceleration_X : ").split(', ')][:samples]
            df['linear_acceleration_X'] = pd.to_numeric(df['linear_acceleration_X'])
            df['linear_acceleration_Y'] = [float(x) for x in input("\nEnter 128 samples for linear_acceleration_Y : ").split(', ')][:samples]
            df['linear_acceleration_Y'] = pd.to_numeric(df['linear_acceleration_Y'])
            df['linear_acceleration_Z'] = [float(x) for x in input("\nEnter 128 samples for linear_acceleration_Z : ").split(', ')][:samples]
            df['linear_acceleration_Y'] = pd.to_numeric(df['linear_acceleration_Y'])
            series_id=[i for j in range(128)]
            df['series_id']=series_id
            df['series_id'] = pd.to_numeric(df['series_id'])
            
            data=pd.concat([data,df])
    
    def quaternion_to_euler(x, y, z, w): # function to convert orientation to euler angles
            
            t0 = +2.0 * (w * x + y * z)
            t1 = +1.0 - 2.0 * (x * x + y * y)
            X = math.atan2(t0, t1)

            t2 = +2.0 * (w * y - z * x)
            t2 = +1.0 if t2 > +1.0 else t2
            t2 = -1.0 if t2 < -1.0 else t2
            Y = math.asin(t2)

            t3 = +2.0 * (w * z + x * y)
            t4 = +1.0 - 2.0 * (y * y + z * z)
            Z = math.atan2(t3, t4)

            return X, Y, Z

    ###########################################################################################################
    x=list(data.orientation_X.values)  ## adding additional features euler angles and total(resultant)
    y=list(data.orientation_Y.values)
    z=list(data.orientation_Z.values)
    w=list(data.orientation_W.values)
    e_x=[]
    e_y=[]
    e_z=[]

    for i in range(0,len(x)):
            a,b,c=quaternion_to_euler(x[i], y[i], z[i], w[i])
            e_x.append(a)
            e_y.append(b)
            e_z.append(c)
    data['euler_x']=e_x
    data['euler_x'] = pd.to_numeric(data['euler_x'])
    data['euler_y']=e_y
    data['euler_y'] = pd.to_numeric(data['euler_y'])
    data['euler_z']=e_z
    data['euler_z'] = pd.to_numeric(data['euler_z'])

    data['total_angular_velocity'] = (data['angular_velocity_X'] ** 2 + data['angular_velocity_Y'] ** 2 + data['angular_velocity_Z'] ** 2) ** 0.5
    data['total_linear_acceleration'] = (data['linear_acceleration_X'] ** 2 + data['linear_acceleration_Y'] ** 2 + data['linear_acceleration_Z'] ** 2) ** 0.5
    #####################################################################################################################   

    def perform_feature_engineering(df):  ## function to extract features from all signals
            df_out = pd.DataFrame()
            
            def mean_change_of_abs_change(x):
                return np.mean(np.diff(np.abs(np.diff(x))))

            def mean_abs_change(x):
                return np.mean(np.abs(np.diff(x)))
            
            for col in (df.columns):  # for each columns we will extracts
                if col != 'series_id':
                    
                    df_out[col + '_mean'] = df.groupby(['series_id'])[col].mean() # mean value of 128 time stamps

                    
                    df_out[col + '_min'] = df.groupby(['series_id'])[col].min() #min value
                    
                    df_out[col + '_max'] = df.groupby(['series_id'])[col].max() # maximum value
                    
                    df_out[col + '_std'] = df.groupby(['series_id'])[col].std() # std
                    
                    df_out[col + '_mad'] = df.groupby(['series_id'])[col].mad()  ## median absolute deviation
                    
                    df_out[col + '_med'] = df.groupby(['series_id'])[col].median() # median 
                    
                    df_out[col + '_skew'] = df.groupby(['series_id'])[col].skew() # skewness
                    
                    df_out[col + '_range'] = df_out[col + '_max'] - df_out[col + '_min'] ## range max-min
                    
                    df_out[col + '_max_to_min'] = df_out[col + '_max'] / df_out[col + '_min'] ## ratio max/min
                    
                    df_out[col + '_mean_abs_change'] = df.groupby('series_id')[col].apply(mean_abs_change) ## mean absolute change
                    
                    df_out[col + '_mean_change_of_abs_change'] = df.groupby('series_id')[col].apply(mean_change_of_abs_change) ## mean chaange of absolute change/diff
                    
                    df_out[col + '_abs_max'] = df.groupby('series_id')[col].apply(lambda x: np.max(np.abs(x))) ## maximum of absolute values
                    
                    df_out[col + '_abs_min'] = df.groupby('series_id')[col].apply(lambda x: np.min(np.abs(x))) ## minimum of absolute values
                    
                    df_out[col + '_abs_mean'] = df.groupby('series_id')[col].apply(lambda x: np.mean(np.abs(x))) ## mean of absolute values
                    
                    df_out[col + '_abs_std'] = df.groupby('series_id')[col].apply(lambda x: np.std(np.abs(x))) ## std of absolute values
                    
                    df_out[col + '_abs_avg'] = (df_out[col + '_abs_min'] + df_out[col + '_abs_max'])/2 ## average of absolute min and absolute max
                    
                    df_out[col + '_abs_range'] = df_out[col + '_abs_max'] - df_out[col + '_abs_min'] ## diff of absolute max and absolute min values

            return df_out
    set_=perform_feature_engineering(data)
    min=MinMaxScaler()
    set_=pd.DataFrame(min.fit_transform(set_),columns=set_.columns)

    test_data=pd.DataFrame()
    for i in range(0,3):
        if i==0:
            model = pickle.load(open('XGB_prob.sav', 'rb'))
        elif i==1:
            model=pickle.load(open('RF_prob.sav','rb'))
        elif i==2:
            model=pickle.load(open('NB_prob.sav','rb'))
        pred=pd.DataFrame(model.predict_proba(np.array(set_)))
        test_data=pd.concat([test_data,pred],axis=1)
    model=pickle.load(open('XGB_final.sav','rb'))
    result=model.predict(np.array(test_data))

    print('***************************************************************************************************************************************************')
    return print('Predictions are {}'.format(result))
    
```

```python
def performance_measure():

    ### this function is exactly the same as previous one, here we are taking class labels as well and at the end instead displaying class we are printing accuracy
    n=int(input('Please Enter Number of data points you want to predict'))
    data=pd.DataFrame(columns=['orientation_X','orientation_Y','orientation_Z','orientation_W','angular_velocity_X','angular_velocity_Y','angular_velocity_Z','linear_acceleration_X','linear_acceleration_Y','linear_acceleration_Z','series_id'])
    classes=[]
    for i in range(n):
            print('Enter values for datapoint number {}'.format(i+1))
            samples= 128
            df=pd.DataFrame()
            df['orientation_X'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_X : ").split(', ')][:samples]
            df['orientation_X'] = pd.to_numeric(df['orientation_X'])
            df['orientation_Y'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_Y :").split(', ')][:samples]
            df['orientation_Y'] = pd.to_numeric(df['orientation_Y'])
            df['orientation_Z'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_Z :").split(', ')][:samples]
            df['orientation_Z'] = pd.to_numeric(df['orientation_Z'])
            df['orientation_W'] = [float(x) for x in input("\nEnter 128 samples for ORIENTATION_W :").split(', ')][:samples]
            df['orientation_W'] = pd.to_numeric(df['orientation_W'])

            df['angular_velocity_X'] = [float(x) for x in input("\nEnter 128 samples for angular_velocity_X : ").split(', ')][:samples]
            df['angular_velocity_X'] = pd.to_numeric(df['angular_velocity_X'])
            df['angular_velocity_Y'] = [float(x) for x in input("\nEnter 128 samples for angular_velocity_Y : ").split(', ')][:samples]
            df['angular_velocity_Y'] = pd.to_numeric(df['angular_velocity_Y'])
            df['angular_velocity_Z'] = [float(x) for x in input("\nEnter 128 samples for angular_velocity_Z : ").split(', ')][:samples]
            df['angular_velocity_Z'] = pd.to_numeric(df['angular_velocity_Z'])

            df['linear_acceleration_X'] = [float(x) for x in input("\nEnter 128 samples for linear_acceleration_X : ").split(', ')][:samples]
            df['linear_acceleration_X'] = pd.to_numeric(df['linear_acceleration_X'])
            df['linear_acceleration_Y'] = [float(x) for x in input("\nEnter 128 samples for linear_acceleration_Y : ").split(', ')][:samples]
            df['linear_acceleration_Y'] = pd.to_numeric(df['linear_acceleration_Y'])
            df['linear_acceleration_Z'] = [float(x) for x in input("\nEnter 128 samples for linear_acceleration_Z : ").split(', ')][:samples]
            df['linear_acceleration_Y'] = pd.to_numeric(df['linear_acceleration_Y'])
            series_id=[i for j in range(128)]
            df['series_id']=series_id
            df['series_id'] = pd.to_numeric(df['series_id'])
            
            data=pd.concat([data,df])
            #input class label
            classes.append(str(input('Please Enter surface type for this data point, options are : \n"carpet",\n"concrete",\n"fine_concrete",\n"hard_tiles",\n"hard_tiles_large_space",\n"soft_pvc",\n"soft_tiles",\n"tiled",\n"wood"')))
    
            
    def quaternion_to_euler(x, y, z, w):
            
            t0 = +2.0 * (w * x + y * z)
            t1 = +1.0 - 2.0 * (x * x + y * y)
            X = math.atan2(t0, t1)

            t2 = +2.0 * (w * y - z * x)
            t2 = +1.0 if t2 > +1.0 else t2
            t2 = -1.0 if t2 < -1.0 else t2
            Y = math.asin(t2)

            t3 = +2.0 * (w * z + x * y)
            t4 = +1.0 - 2.0 * (y * y + z * z)
            Z = math.atan2(t3, t4)

            return X, Y, Z
    x=list(data.orientation_X.values)
    y=list(data.orientation_Y.values)
    z=list(data.orientation_Z.values)
    w=list(data.orientation_W.values)
    e_x=[]
    e_y=[]
    e_z=[]

    for i in range(0,len(x)):
            a,b,c=quaternion_to_euler(x[i], y[i], z[i], w[i])
            e_x.append(a)
            e_y.append(b)
            e_z.append(c)
    data['euler_x']=e_x
    data['euler_x'] = pd.to_numeric(data['euler_x'])
    data['euler_y']=e_y
    data['euler_y'] = pd.to_numeric(data['euler_y'])
    data['euler_z']=e_z
    data['euler_z'] = pd.to_numeric(data['euler_z'])

    data['total_angular_velocity'] = (data['angular_velocity_X'] ** 2 + data['angular_velocity_Y'] ** 2 + data['angular_velocity_Z'] ** 2) ** 0.5
    data['total_linear_acceleration'] = (data['linear_acceleration_X'] ** 2 + data['linear_acceleration_Y'] ** 2 + data['linear_acceleration_Z'] ** 2) ** 0.5
       

    def perform_feature_engineering(df):
            df_out = pd.DataFrame()
            
            def mean_change_of_abs_change(x):
                return np.mean(np.diff(np.abs(np.diff(x))))

            def mean_abs_change(x):
                return np.mean(np.abs(np.diff(x)))
            
            for col in (df.columns):  # for each columns we will extracts
                if col != 'series_id':
                    
                    df_out[col + '_mean'] = df.groupby(['series_id'])[col].mean() # mean value of 128 time stamps

                    
                    df_out[col + '_min'] = df.groupby(['series_id'])[col].min() #min value
                    
                    df_out[col + '_max'] = df.groupby(['series_id'])[col].max() # maximum value
                    
                    df_out[col + '_std'] = df.groupby(['series_id'])[col].std() # std
                    
                    df_out[col + '_mad'] = df.groupby(['series_id'])[col].mad()  ## median absolute deviation
                    
                    df_out[col + '_med'] = df.groupby(['series_id'])[col].median() # median 
                    
                    df_out[col + '_skew'] = df.groupby(['series_id'])[col].skew() # skewness
                    
                    df_out[col + '_range'] = df_out[col + '_max'] - df_out[col + '_min'] ## range max-min
                    
                    df_out[col + '_max_to_min'] = df_out[col + '_max'] / df_out[col + '_min'] ## ratio max/min
                    
                    df_out[col + '_mean_abs_change'] = df.groupby('series_id')[col].apply(mean_abs_change) ## mean absolute change
                    
                    df_out[col + '_mean_change_of_abs_change'] = df.groupby('series_id')[col].apply(mean_change_of_abs_change) ## mean chaange of absolute change/diff
                    
                    df_out[col + '_abs_max'] = df.groupby('series_id')[col].apply(lambda x: np.max(np.abs(x))) ## maximum of absolute values
                    
                    df_out[col + '_abs_min'] = df.groupby('series_id')[col].apply(lambda x: np.min(np.abs(x))) ## minimum of absolute values
                    
                    df_out[col + '_abs_mean'] = df.groupby('series_id')[col].apply(lambda x: np.mean(np.abs(x))) ## mean of absolute values
                    
                    df_out[col + '_abs_std'] = df.groupby('series_id')[col].apply(lambda x: np.std(np.abs(x))) ## std of absolute values
                    
                    df_out[col + '_abs_avg'] = (df_out[col + '_abs_min'] + df_out[col + '_abs_max'])/2 ## average of absolute min and absolute max
                    
                    df_out[col + '_abs_range'] = df_out[col + '_abs_max'] - df_out[col + '_abs_min'] ## diff of absolute max and absolute min values

            return df_out
    set_=perform_feature_engineering(data)
    min=MinMaxScaler()
    set_=pd.DataFrame(min.fit_transform(set_),columns=set_.columns)
    
    test_data=pd.DataFrame()
    for i in range(0,3):
        if i==0:
            model = pickle.load(open('XGB_prob.sav', 'rb'))
        elif i==1:
            model=pickle.load(open('RF_prob.sav','rb'))
        elif i==2:
            model=pickle.load(open('NB_prob.sav','rb'))
        pred=pd.DataFrame(model.predict_proba(np.array(set_)))
        test_data=pd.concat([test_data,pred],axis=1)
    model=pickle.load(open('XGB_final.sav','rb'))
    result=model.predict(np.array(test_data))
    print('***************************************************************************************************************************************************')
    return print('Accuracy score for this run is: {}'.format(accuracy_score(classes,result)))
```
